```python
import json
import collections
import pandas as pd
```

```python
with open('../data/QA/0_train.json','r') as f:
    data = f.readlines()
with open('../data/QA/1_train.json','r') as f:
    data1 = f.readlines()
with open('../data/QA/0_test.json','r') as f:
    data2 = f.readlines()
with open('../data/QA/1_test.json','r') as f:
    data3 = f.readlines()
```

```python
all_data = data+data1+data2+data3
```

```python
all_data = [json.loads(i) for i in all_data]
```

```python
statement = [i['statement'] for i in all_data]
option = ['[SEP]'.join(i['option_list'].values()) for i in all_data]
```

```python
statement_len = [len(i) for i in statement]
option_len = [len(i) for i in option]
```
